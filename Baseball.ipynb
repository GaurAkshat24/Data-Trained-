```python
from sklearn.preprocessing import StandardScaler

cols = list(df.columns)

std_scaler = StandardScaler()

scaled_df = std_scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=cols)

cols = list(scaled_df.columns)
fig, ax = plt.subplots(figsize=(20, 20))

for col in cols:
    sns.kdeplot(scaled_df[col], ax=ax, label=col)
    ax.set_title('After Standard Scaler')
    ax.set_xlabel('Scale')
    ax.legend(loc=1)
    
plt.show()
```

```python
import statsmodels.api as sm

df = pd.concat([df['R'], scaled_df], axis=1)

x = df.drop(['R'], axis=1)
x = sm.add_constant(x)
y = df['R']


lm = sm.OLS(y, x)
result = lm.fit()
```

```python
from sklearn import linear_model, metrics
from sklearn.linear_model import LinearRegression

x = np.array(df['W']).reshape(-1, 1)
y = df['W']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

lm = linear_model.LinearRegression().fit(x_train, y_train)
y_predict = lm.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score

x = np.array(df['W']).reshape(-1, 1)
y = df['W']

model = LinearRegression()

cv_r2 = cross_val_score(model, x, y, scoring='r2', cv=10)
cv_mse = cross_val_score(model, x, y, scoring='neg_mean_squared_error', cv=10)
cv_rmse = np.sqrt(-1 * cv_mse)
```
