```python
from sklearn.linear_model import LogisticRegression
```

```python
for i in range (1,200):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=i)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
pre=lr.predict(x_test)
```

```python
acc=accuracy_score(y_test,pre)
```

```python
maxacc=0
maxrs=0
if acc>maxacc:
    maxacc=acc
    maxrs=i
```

```python
print('Best accuracy is',maxacc,'on Random State',maxrs)
```

```python
print(accuracy_score(y_test,pre))
print(confusion_matrix(y_test,pre))
print(classification_report(y_test,pre))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
pre=dt.predict(x_test)
```

```python
print(accuracy_score(y_test,pre))
print(confusion_matrix(y_test,pre))
print(classification_report(y_test,pre))
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
pre=svc.predict(x_test)
print(accuracy_score(y_test,pre))
print(confusion_matrix(y_test,pre))
print(classification_report(y_test,pre))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
pre=rf.predict(x_test)
print(accuracy_score(y_test,pre))
print(confusion_matrix(y_test,pre))
print(classification_report(y_test,pre))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cvr=cross_val_score(lr,x,y,cv=5)
print('Cross validation score of Logistic Regression',cvr.mean())
```

```python
cvr=cross_val_score(dt,x,y,cv=5)
print('Cross validation score of Decision Tree',cvr.mean())
```

```python
cvr=cross_val_score(svc,x,y,cv=5)
print('Cross validation score of SVC',cvr.mean())
```

```python
cvr=cross_val_score(rf,x,y,cv=5)
print('Cross validation score of Random Forest',cvr.mean())
```
